```python
# 필요한 라이브러리 임포트
import numpy as np

import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.efficientnet_v2 import preprocess_input
import matplotlib.pyplot as plt
```

```python
# 데이터 경로 설정
data_path = "/home/ec2-user/SageMaker/Downloaded_data/data/다이캐스팅/g2/"  # g2 폴더 경로

# g2 폴더 내의 데이터셋 확인
folders = os.listdir(data_path)
print("Available datasets:", folders)
```

```python
# 사용할 데이터셋 폴더 선택
selected_folder = os.path.join(data_path, "dataset_2023-10-17")
print("Selected folder:", selected_folder)
```

```python
# 이미지 파일 목록 가져오기
image_files = [f for f in os.listdir(selected_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]
print("Found image files:", image_files[:5])  # 처음 5개 이미지 출력
```

```python
# EfficientNetV2에 적합한 전처리 함수
def preprocess_image(image_path, target_size=(224, 224)):
    # 이미지 로드
    image = tf.io.read_file(image_path)
    image = tf.image.decode_png(image, channels=3)  # PNG 이미지 디코딩
    
    # 크기 조정
    image = tf.image.resize(image, target_size)
    
    # 픽셀 값을 [-1, 1]로 정규화
    image = preprocess_input(image)
    return image
```

```python
# 첫 번째 이미지 경로
test_image_path = os.path.join(selected_folder, image_files[0])
print("Test image path:", test_image_path)

# 전처리 실행
processed_image = preprocess_image(test_image_path)

# 결과 출력
print("Processed image shape:", processed_image.shape)
```

```python
# 선택한 폴더에 어떤 파일이 있는지 확인
print("Selected folder:", selected_folder)
print("Files in folder:", os.listdir(selected_folder))
```

```python
# 하위 폴더의 이미지 파일도 포함
for root, dirs, files in os.walk(selected_folder):
    print(f"Searching in: {root}")
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            print(f"Found image: {os.path.join(root, file)}")
```

```python
# 원본 이미지 로드
original_image = tf.image.decode_png(tf.io.read_file(test_image_path), channels=3)

# 시각화
plt.figure(figsize=(10, 5))

# 원본 이미지
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(original_image.numpy().astype("uint8"))

# 전처리된 이미지
plt.subplot(1, 2, 2)
plt.title("Preprocessed Image")
plt.imshow((processed_image.numpy() + 1) / 2)  # [-1, 1]을 [0, 1]로 변환
plt.show()
```

```python
def preprocess_batch(folder_path, target_size=(224, 224)):
    processed_images = []
    image_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

    for image_path in image_paths:
        processed_image = preprocess_image(image_path, target_size)
        processed_images.append(processed_image.numpy())
    
    return np.array(processed_images)

# 전체 이미지 전처리 실행
processed_data = preprocess_batch(selected_folder)
print("Processed data shape:", processed_data.shape)
```

```python
# Numpy 데이터로 저장
output_file = "/home/ec2-user/SageMaker/processed_g2_data.npy"
np.save(output_file, processed_data)
print(f"Processed data saved to {output_file}")
```

```python
# 저장된 Numpy 파일 로드
loaded_data = np.load(output_file)
print("Loaded data shape:", loaded_data.shape)
```
